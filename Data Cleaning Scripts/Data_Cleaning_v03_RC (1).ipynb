```python
# dataframe where data from fitabase is called into
df = pd.read_csv('sleepDay_merged2.csv')
```

```python
# dataframe with the consent dates of each participant
df_C = pd.read_excel('Sp23_consent_dates.xlsx')
#df_C
```

```python
dfC_sorted = df_C.sort_values('ID')
#dfC_sorted
```

```python
df_C
```

```python
# pulls specific column desired
unique_ids = df['Id'].unique()
dfs={}
dfs_consent = {}
for id in unique_ids:
    curr_df = df.query("Id == @id")[["SleepDay", "TotalTimeInBed"]]
    curr_df["SleepDay"] = pd.to_datetime(curr_df["SleepDay"])

    consent_df = dfC_sorted.query("ID ==@id")[["Consent Date"]]
    consent_df["Consent Date"] = pd.to_datetime(consent_df["Consent Date"])
    dfs[id] = curr_df
    dfs[id].reset_index(drop=True, inplace=True)
    dfs_consent[id] = consent_df
```

```python
# calculates the number of days the first data point is after the consent date
for id in unique_ids:
    consent_date = dfs_consent[id]["Consent Date"].iloc[0]
    
    for  i in range(dfs[id]["SleepDay"].size):
        activity_date = dfs[id]["SleepDay"].iloc[i]
        dfs[id]["SleepDay"].iloc[i] = activity_date - consent_date
        days_int = int(dfs[id]["SleepDay"][i].days)
        dfs[id]["SleepDay"].iloc[i] = days_int
        
        
```

```python
#for id in unique_ids:
 #   empty_rows1 = pd.DataFrame(index=range(100 - dfs1[id].shape[0]))
  #  dfs[id] = pd.concat([dfs1[id], empty_rows1], ignore_index=True)
   # dfs[id]=dfs1[id].shift(dfs1[id]["Day"][0])
    #print(dfs1)
```

```python
# makes all dfs have 100 rows
for id in unique_ids:
    empty_rows = pd.DataFrame(index=range(120 - dfs[id].shape[0]))
    dfs[id] = pd.concat([dfs[id], empty_rows], ignore_index=True)
    dfs[id] = dfs[id].shift(dfs[id]["SleepDay"][0])
    dfs[id].fillna(0, inplace=True)
```

```python
# retrieves only the column with the data desired
for id in unique_ids:
    curr = dfs[id] ["TotalTimeInBed"]
    dfs[id] = curr
    #curr_pa = dfs2[id] #["TotalTimeInBed"]
    #dfs2[id] = curr_pa
# combines all the dfs in the dictionary into one df
combined_df = pd.concat(dfs, axis=1)
combined_df
```

```python
# replaces any NaN values with 0
combined_df.fillna(0, inplace=True)
#combined_df_pa.fillna(0, inplace=True)
```

```python
# name of new file that would be created/appended
file_name = 'SP23_dayTimeWearing2.xlsx'
```

```python
# Check if the file exists, and create it if it doesn't
try:
    with pd.ExcelWriter(file_name, mode='a') as writer:
        combined_df.to_excel(writer, sheet_name='TotalTimeInBed', index=True)
except FileNotFoundError:
    # If the file doesn't exist, create it with the data
    with pd.ExcelWriter(file_name, mode='w') as writer:
        combined_df.to_excel(writer, sheet_name='TotalTimeInBed', index=True)
```
